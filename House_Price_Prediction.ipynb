```python
dummy = pd.get_dummies(data['furnishingstatus'], drop_first = True)
dummy = dummy.astype(int)
dummy
```

```python
data = pd.concat([data,dummy],axis = 1)
```

```python
data.drop('furnishingstatus', axis = 1, inplace = True)
```

```python
data
```

```python
x = data.drop('price', axis = 1)
y = data['price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
scale = 'area bedrooms bathrooms stories parking '.split()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
x_train[scale] = scaler.fit_transform(x_train[scale])
```

```python
x_train
```

```python
x_test[scale] = scaler.transform(x_test[scale])
x_test
```

```python
x_train[scale].std()
```

```python
label = 'mainroad guestroom basement hotwaterheating airconditioning prefarea'.split()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in label:
  x_train[i] = le.fit_transform(x_train[i])
  x_test[i] = le.fit_transform(x_test[i])
```

```python
x_train
```

```python
x_test
```

```python
print(x_train.std())
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python

pred = model.predict(x_test)
pred
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,root_mean_squared_error,mean_squared_error
print(mean_absolute_error(y_test,pred))
print(mean_squared_error(y_test,pred))
print(np.sqrt(mean_squared_error(y_test,pred)))
print(r2_score(y_test,pred))
```
